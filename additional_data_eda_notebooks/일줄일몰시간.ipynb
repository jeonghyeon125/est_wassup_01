```python
import pandas as pd
import numpy as np
```

```python
import requests
import xmltodict

url = 'http://apis.data.go.kr/B090041/openapi/service/RiseSetInfoService/getAreaRiseSetInfo'
params ={
  'serviceKey' : '3Su2cNlpABX0x35gWBwcNFDzP2Qul8d6VcT4UB3M+o/OQwtBzqHm4p7wvLguJsKKpDG7FN8ZYjOKfjyt7aNezQ==',
  'locdate' : '2019',
  'location' : '대구'
}

raw_records = []
sun_record = []
for year in '2019 2020 2021'.split():
  for month in '01 02 03 04 05 06 07 08 09 10 11 12'.split():
    params['locdate'] = year+month
    response = requests.get(url, params=params)
    raw_records.append(response.content)

for res in raw_records:
  records = xmltodict.parse(res)['response']['body']['items']['item']
  for r in records:
    sun_record.append({
        'month': r['locdate'][4:6],
        'sunrise': r['sunrise'][:2],
        'sunset': r['sunset'][:2]
      })
```

```python
sun_record = pd.DataFrame(sun_record)
sun_record['sunrise'] = sun_record['sunrise'].astype(int)
sun_record['sunset'] = sun_record['sunset'].astype(int)
sun_record
```

```python
sun_record = sun_record.groupby('month').mean().reset_index()
sun_record
```

```python
sun_record.to_csv('../../main/data/origin/additionals/sun_records_monthly_mean.csv')
```
