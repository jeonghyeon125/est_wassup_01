```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
moving_by_age = pd.read_csv('./externals/대구_빅데이터_활용센터/6_1_moving_by_age.csv', encoding='cp949', index_col=0)
moving_by_days = pd.read_csv('./externals/대구_빅데이터_활용센터/6_1_moving_by_days.csv', encoding='cp949', index_col=0)
moving_by_times = pd.read_csv('./externals/대구_빅데이터_활용센터/6_1_moving_by_times.csv', encoding='cp949', index_col=0)
```

```python
df_moving = {
  'age': moving_by_age,
  'time': moving_by_times,
  'days': moving_by_days
}
```

```python
for k, v in df_moving.items():
  new_columns = [col+'_sum' if col.count('.') == 0 else col.split('.')[0]+'_count' for col in v.columns]
  v.columns = new_columns
  v.reset_index(drop=True, inplace=True)
  v.drop(0, inplace=True)
  df_moving[k] = v
  
```

# 월별 나이별 법정동별 유동인구
- 컬럼병 수정
- 평균값 도출

```python
moving_by_age.columns = [col.replace('WMAN_','').replace('MAN_', '') for col in moving_by_age.columns]
moving_by_age
```

```python
for k,v in df_moving.items():
  new_v = v.iloc[:,4:].map(float)
  df_moving[k] = pd.concat([v.iloc[:,:4], new_v], axis=1)
```

# 평균값 계산
- 동멸 연령별 월별 평균값 계산

```python
moving_by_age_sum = df_moving['age']
moving_by_age_sum = moving_by_age_sum.iloc[:, 4:16:2] + moving_by_age_sum.iloc[:, 16::2]

moving_by_age_sum['구'] = df_moving['age']['SGNG_NM_sum']
moving_by_age_sum['year'] = df_moving['age']['STD_YM_sum']
moving_by_age_sum['동'] = df_moving['age']['LDONG_NM_sum']
moving_by_age_sum.columns = ['10대', '20대', '30대', '40대', '50대', '60대', '구', '년', '동']
moving_by_age_sum['월'] = moving_by_age_sum['년'].astype(str).str[4:6]
moving_by_age_sum = moving_by_age_sum.drop('년', axis=1).groupby(['월','구','동']).sum()
moving_by_age_sum
```

```python
moving_by_age_cnt = df_moving['age']
moving_by_age_cnt = moving_by_age_cnt.iloc[:, 5:17:2] + moving_by_age_cnt.iloc[:, 17::2]
moving_by_age_cnt['year'] = df_moving['age']['STD_YM_sum']
moving_by_age_cnt['구'] = df_moving['age']['SGNG_NM_sum']
moving_by_age_cnt['동'] = df_moving['age']['LDONG_NM_sum']
moving_by_age_cnt.columns = ['10대', '20대', '30대', '40대', '50대', '60대', '년', '구', '동']
moving_by_age_cnt['월'] = moving_by_age_cnt['년'].astype(str).str[4:6]
moving_by_age_cnt = moving_by_age_cnt.drop('년', axis=1).groupby(['월','구','동']).sum()
moving_by_age_cnt
```

```python
moving_by_age_mean = moving_by_age_sum / moving_by_age_cnt
moving_by_age_mean
```

```python
moving_by_age_mean_12 = ((moving_by_age_mean.loc['01'] + moving_by_age_mean.loc['11'])/2)
moving_by_age_mean_12 = moving_by_age_mean_12.reset_index()
moving_by_age_mean_12['월'] = '12'
moving_by_age_mean_12 = moving_by_age_mean_12.groupby(['월', '구', '동']).sum()
moving_by_age_mean_12
```

```python
moving_by_age_mean = pd.concat([moving_by_age_mean, moving_by_age_mean_12])
moving_by_age_mean
```

```python
df_moving['age'] = moving_by_age_mean
df_moving['age']
```

# 동의 면적별 비율 구하기
- 기존 데이터 / 면적

```python
area_size = pd.read_csv('../../main/data/origin/additionals/jiri_dong_area_size.csv', index_col=0)
area_size
```

```python
def calc_mildo_by_area(row: pd.Series):
  row = {**row}
  area = area_size.loc[area_size.iloc[:,0] == row['동']]['area'].iloc[0]
  for k,v in row.items():
    if type(v) == float:
      row[k] = row[k]/area
  return pd.Series(row)

for k,v in df_moving.items():
  df_moving[k] = v.reset_index().apply(
    lambda row: calc_mildo_by_area(row),
    axis=1
  )
```
