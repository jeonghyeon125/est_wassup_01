```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
weather = pd.read_excel('./externals/공공_데이터_포털/기상개황_20231204115303.xlsx')
weather.head()
```

```python
weather = weather.drop([1])
weather
```

```python
weather.columns = weather.iloc[0]
weather
```

```python
weather = weather.drop(0).reset_index(drop=True).T
weather.columns = '1월 2월 3월 4월 5월 6월 7월 8월 9월 10월 11월 12월'.split()
weather
```

```python
weather = weather.reset_index()
weather
```

```python
weather = weather.map(lambda x: np.nan if x == '-' or x =='...' else x).dropna().drop(0).set_index(0).map(float).reset_index().groupby([0]).mean().T.drop(columns=[
  '평균해면기압 (hPa)', '최고극값기온 (℃)', '최대순간풍속 (m/s)', '최소상대습도 (%)', '최저극값기온 (℃)', '평균풍속 (m/s)'
])
weather
```

```python
weather.to_csv('../../main/data/origin/additionals/weather_overall_monthly.csv')
```
