```python
jiri_dong = gpd.read_file(jiri_dong_path)
jiri_gu = gpd.read_file(jiri_gu_path).iloc[:, 2:]
```

```python
df_jiri = {
  'dong': jiri_dong,
  'gu': jiri_gu
}
```

```python
def find_gu(geometry, name):
  dong = []
  for idx in df_jiri[name].index:
    if name == 'gu':
      if geometry.within(df_jiri[name].iloc[idx]['geometry']):
        dong.append(df_jiri[name].iloc[idx, 0])
    else:
      if geometry.intersects(df_jiri[name].iloc[idx]['geometry']):
        dong.append(df_jiri[name].iloc[idx, 0])
  return ','.join(dong)
```

```python
for area in jiri_dong.keys():
  for acc in df_accidents.keys():
    df_accidents[acc] = df_accidents[acc].apply(
      lambda row: pd.Series({**row, 'dong': find_gu(row['geometry'], 'dong')}),
      axis=1
    )
    df_accidents[acc] = df_accidents[acc].apply(
      lambda row: pd.Series({**row, 'gu': find_gu(row['geometry'], 'gu')}),
      axis=1
    )
```

# 구 보류

```python
def separate_multiple_areas(df: gpd.GeoDataFrame):
  df = df.copy()
  new_gdf = []
  for idx, row in df.iterrows():
    area = row['dong'].split(',')
    if len(area) > 1:
      df.loc[idx, 'dong'] = area[0]
      for i in range(1, len(area)):
        new_gdf.append({**row, 'dong': area[i]})
  return pd.concat([df, pd.DataFrame(new_gdf)]).reset_index(drop=True)
```

```python
for k,v in df_accidents.items():
  new_df = separate_multiple_areas(v)
  df_accidents[k] = new_df
```

```python
for k,v in df_accidents.items():
  df_accidents[k] = v.drop(columns=['geometry'])

df_accidents['bad_pedestrian'].head()
```

```python
for k,v in df_accidents.items():
  df_accidents[k] = v.groupby(['dong']).sum()
```

```python
for k,v in df_accidents.items():
  df_accidents[k] = v.drop(['id', 'count', 'occrrnc_cnt_mean', 'caslt_cnt_mean', 'dth_dnv_cnt_mean', 'se_dnv_cnt_mean', 'sl_dnv_cnt_mean', 'wnd_dnv_cnt_mean', 'wnd_dnv_cnt_sum'], axis=1)\
    .rename(columns={'occrrnc_cnt_sum': '사고건수', 'caslt_cnt_sum': '사상자수', 'dth_dnv_cnt_sum': '사망자수', 'se_dnv_cnt_sum': '중상자수', 'sl_dnv_cnt_sum': '경상자수'}).reset_index()
# occrnc_cnt 사고건수
# caslt_cnt 사상자수
# dth_dnv_cnt 사망자수
# se_dnv_cnt 중상자수
# sl_dnv_cnt 경상자수
# wnd_dnv_cnt 부상신고자수
df_accidents['bad_pedestrian'].head()
```

```python
def calc_mildo_by_area(row: pd.Series):
  row = {**row}
  area = jiri_dong.loc[jiri_dong.iloc[:,0] == row['dong']]['area'].iloc[0]
  for col in ['사고건수',	'사상자수',	'사망자수',	'중상자수',	'경상자수']:
    row[col] = row[col]/area
  return pd.Series(row)

for k,v in df_accidents.items():
  df_accidents[k] = v.apply(
    lambda row: calc_mildo_by_area(row),
    axis=1
  )

df_accidents['bad_pedestrian']
```
